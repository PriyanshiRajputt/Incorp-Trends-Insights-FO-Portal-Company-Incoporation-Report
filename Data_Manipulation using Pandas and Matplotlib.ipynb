```python
import seaborn as sns
```

```python
data=sns.countplot(data=IC_df,x='Class',order=IC_df["Class"].value_counts().index,color="red")
plt.title('Different Classes of Company')
plt.xlabel("Classes")
plt.ylabel("Values")
plt.figure(figsize=(3,4))
for container in data.containers:
    data.bar_label(container)
```

```python
data=sns.countplot(data=IC_df,x='Company Type',order=IC_df["Company Type"].value_counts().index,color="red")
plt.title('Different Company Types ')
plt.xlabel("Company Type")
plt.ylabel("Counts")
plt.xticks(rotation=70)
plt.figure(figsize=(3,4))
for container in data.containers:
    data.bar_label(container)
```
